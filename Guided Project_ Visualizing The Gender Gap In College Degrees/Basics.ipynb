```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_gray = (171/255, 171/255, 171/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

Engineering and Computer Science have a historically large difference in men and women shares. Psychology started as a men dominated field but women have rapidly taken over. Women have conquered Biology as well, though not as forcefully nor as rapidly. In the Physical Sciences there has been a gradual decreasing number of men and increasing number of women, though the trend seems to have halted for a few years. In Math and Statistics, there was a gradual convergence but it has shifted to a slight divergence; men have held a higher share throughout. 

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize = (18, 18))
all_cats = [stem_cats, lib_arts_cats, other_cats]
for i in range(len(all_cats)):
    picked_cats = all_cats[i]
    start = i
    for j in range(len(picked_cats)):
        cat = picked_cats[j]
        sp = start + 3*j + 1
        ax = fig.add_subplot(6, 3, sp)
        ax.plot(women_degrees["Year"], women_degrees[cat], c=cb_dark_blue, label="Women", linewidth=3)
        ax.plot(women_degrees["Year"], 100-women_degrees[cat], c=cb_orange, label="Men", linewidth=3)
        ax.set_title(cat)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.tick_params(bottom="off", top="off", left="off", right="off")
        ax.set_yticks([0, 100])
        ax.axhline(50, c=cb_gray, alpha=0.3)
        if sp < 16 and sp != 14:
            ax.tick_params(labelbottom="off")
        if sp == 1 or sp == 2 or sp == 16 or sp == 3 or sp == 18:
            ax.legend(loc="best")
plt.savefig("gender_degrees.png")
plt.show()
```

In the sciences, Biology, Math, and Physical Sciences have the smallest difference between Men and Women.Out of the six categories, only two--Psychology and Biology--have a higher share of women. Engineering and Psychology have the largest differences. 

In the libarl arts, Social Sciences and History have a near equal number of men and women in the category, though men have a higher share. In the four other categories, however, women hold a marginally higher share. 

In the other categories, Architecture, Business, and Agriculter men hold a higher share but there is a nearly even distribution of men and women. In Health Education, Public Administration, and Education women hold a substantially higher share.
